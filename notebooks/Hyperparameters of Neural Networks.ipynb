```python
import pandas as pd
import numpy as np
import os
```

```python
main_tune_folder = '../tune'
```

```python
def get_most_recent_folder(parent_folder):
    # List all entries in the parent folder
    entries = [os.path.join(parent_folder, entry) for entry in os.listdir(parent_folder)]
    
    # Filter to include only directories
    folders = [entry for entry in entries if os.path.isdir(entry)]

    if not folders:
        return None  # No folders found

    # Find the most recently modified folder
    most_recent_folder = max(folders, key=os.path.getmtime)
    return most_recent_folder
```

```python
tune_name = 'MLP__2025-03-28_19-18-13'
experiment_filename = "experiment_state-2025-03-28_19-18-18"
tune_folder = f"{main_tune_folder}/{tune_name}"
# Read json file as dict
json_file = f"{tune_folder}/{experiment_filename}.json"
with open(json_file, 'r') as file:
    json_data = file.read()
    import json
    json_dict = json.loads(json_data)
```

```python
list_of_trial_results = []
for i in range(len(json_dict['trial_data'])):
    t = json_dict['trial_data'][i][1].replace('\n', '')
    result = json.loads(t)['last_result']
    list_of_trial_results.append(result)
# Convert the list of trial results to a DataFrame
trial_results_df = pd.DataFrame(list_of_trial_results)
trial_results_df.sort_values(by='logQ', inplace=True)
trial_results_df
```

```python
# Display the configuration of the best trial
trial_results_df.iloc[0]['config']
```
