# Load results

```python
test_resuls_folder = '../project_code/debtool_port/tests'
test_date = '20250729'
df = pd.read_csv(f'{test_resuls_folder}/test_init_DEBInitNet_{test_date}.csv', index_col=0)
error_columns = df.columns[4:-1]
df[error_columns] = df[error_columns].astype(bool)
df
```

```python
df['error_message'].value_counts()
```

```python
# Drop rows with no errors
df = df[~df['error_message'].isna()]
# Remove errors from missing data or incompatible model type
# These error messages all contain the string 'DEBInitNet'
df = df[~df['error_message'].str.contains('DEBInitNet')]
# Remove errors from 'z' not being free
df = df[df['error_message'] != "Parameter 'z' should be set free."]
df
```
